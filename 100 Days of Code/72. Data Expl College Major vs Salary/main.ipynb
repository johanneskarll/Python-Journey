### **Lowest Risk Majors**
Kita ingin mencari low-risk major yang mana itu adalah jurusan yang memiliki perbedaan gaji tertinggi dan terendahnya paling kecil. Artinya 10th percentile dan 90th percentilenya paling kecil.

>Pandas memungkinkan kita untuk melakukan spile arithmetic dengan semua kolom, yang perlu kita lakukan adalah menjadi perbedaan antara kedua kolom, terdapat dua cara dibawah ini

```python
# Cara 1 :
clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']

# Cara 2 : (menggunakan subtract())
clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

Output yang dihasilkan akan kita tambahkan menjadi kolom dataframe yang baru dengan cara insert

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary']-clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1,'Spread', spread_col) # karena kita masukin parms 1, jadinya paling awal
clean_df.head()
```

```python
low_risk = clean_df.sort_values("Spread")
low_risk[['Undergraduate Major','Spread']].head()
```

```python
potential_90th = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)
potential_90th[['Undergraduate Major','Spread']].head()
```

```python
# The largest difference betwheen high and low earners
clean_df['Undergraduate Major'][clean_df["Spread"].idxmax()]
```
